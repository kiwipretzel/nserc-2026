```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def commutator(A,B):
    return A@B - B@A




```

```python
rho_0 = np.array([[1, 0], [0, 0]])
h_bar = 1
H = np.array([[0, 1], [1, 0]])
number_of_steps = 200
time_start = 0
time_end = 10
dt = (time_end - time_start) / number_of_steps



rhos = []
times = []
rhos.append(rho_0)
times.append(time_start)



for i in range(1,number_of_steps+1):
    rho_dot = -1j / h_bar * commutator(H, rhos[i-1])
    rhos.append(rhos[i-1]+ rho_dot*dt)
    times.append((float(i*dt) + time_start))
    
#print(rhos)
#print(times)

rhos = np.array(rhos)
times = np.array(times)


plt.plot(times, rhos[:, 0, 0], label="numerical")
plt.plot(times, np.cos(times)**2, label="analytical")
plt.legend()
plt.title("Prob of being in |1>")
plt.xlabel("Time (s)")
plt.ylabel("Prob")
```
